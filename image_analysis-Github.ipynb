```python
import cv2
import glob

images=glob.glob("*.jpg")

for image in images:
    print(image)
    img=cv2.imread(image,0)
    re=cv2.resize(img,(100,100))
    cv2.imshow("Hey",re)
    cv2.waitKey(5000)
    cv2.destroyAllWindows()
    cv2.imwrite("resized_"+image,re)
```

```python
import cv2
```

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalcatface_extended_download.xml')  
```

```python
img=cv2.imread('cat4.jpg')
```

```python
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

faces=face_cascade.detectMultiScale(gray_image,scaleFactor=1.01,minNeighbors=10)

for x,y,w,h in faces:
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
#print(type(faces))
#print(faces)

```

```python
cv2.imwrite('Cat_detect.jpg',img)
cv2.imshow("Grey",img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
import cv2,time
video=cv2.VideoCapture(0)
check, frame=video.read()     #this is the first frame
print(check)
print(frame)
time.sleep(3)

cv2.imshow('capture',frame)
cv2.waitKey(10000)
video.release()
cv2.destroyAllWindows()
```

```python
import cv2,time
video=cv2.VideoCapture(0)   # from camera 0   creating an object like a canvas figure, or video file path
a=1
while True:
    a=a+1
    check, frame=video.read()     #this is the first frame
    print(check)
    print(frame)
    #time.sleep(3)

    cv2.imshow('capture',frame)
    key=cv2.waitKey(1)             # lower number more snmoth
    if key==ord('q'):
        break
    
video.release()
cv2.destroyAllWindows()
print(a)
```
